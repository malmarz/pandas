# How to post your slides to Github

## First, hosting your project on github
### You need to perform the following steps the first time you create a project, you don't have to repeat them:
1. Signup for a github account on github.com. It's free.
2. Create a new repository to store your notebook and files to:
![](imgs/create_repo.png)
3. Give the repository a name, which will be your project name. Then put a check mark near **create a readme file**, and choose **python** for the add .gitignore file at the bottom. Then create the repository:
![](imgs/repo.png)
4. Clone the repository to your computer by selecting **clone or download** on the right:
![](imgs/clone1.png)
5. Select use https, then copy the URL using CTRL-C:
![](imgs/clone2.png)
6. Now in the terminal, create a new directory called **projects** by typing the following in terminal or CMD:
```bash
mkdir projects
```
This will be the directory where you store all your projects in.
7. cd into the projects directory so you can work from it by typing:
```bash
cd projects
```
8. Now clone your github project by typing:
```bash
git clone [!!paste the url of your github repository here!!]
```
9. You will have a new directory that has the same name as your github repository. This directory will be linked with github. Go ahead and place your notebook into that directory.
10. Download and unzip [reveal.js](https://github.com/downloads/hakimel/reveal.js/reveal.js-2.1.zip) into your repository. It should be with your jupyter notebook.
11. cd into the new directory by typing:
```bash
cd [the name of your directory here]
```
12. Once in the directory we want git to store the new files, type:
```bash
git add .
git commit -a -m 'saving my project'
git push origin master
```
13. Now go back to the browser, and refresh your repository. You should see all your files there.


### Now every time you make changes to your project, just type the following to store it on github:
```bash
git add .
git commit -a -m 'saving my project'
git push
```

# Setting up github to show your slides
1. Select settings on your github repository:
![](imgs/settings.png)
2. Scroll down in the settings page and look for **github pages** and make the selection **master branch**, then save:
![](imgs/pages.png)
3. After saving the changes, github will tell you where your website will be published, wait a few minutes because it might not be ready right away then click the link (dont forget to write it down):
![](imgs/publish.png)

### Notice the page that is show is the README.md file.
Open the file and edit it. Put anything you want. The file accepts markdown formatting.
Create an intro page about your project using markdown on your computer then save it.
After saving, to publish, type the following in terminal:

```bash
git add .
git commit -a -m 'saving my project'
git push
```

And you will see the new changes you made on your website

# Preparing and Publishing your slides
1. Prepare your slides by openning your notebook and selecting View -> Cell Tool bar -> Slide show:
![](imgs/cells.png)
2. You will notice an option will appear to the right of each cell, to make the cell a new slide, just select slide:
![](imgs/slides.png)
3. After selecting how your presentation will be arrange by properly slide type for each cell, save your jupyter notebook.
4. Go to terminal or CMD and type the following inside your repository directory:

```bash
jupyter nbconvert your_notebook_name_here.ipynb --to slides --reveal-prefix reveal.js

```
5. Now that the slides are created, publish the files to github as you normally would:
```bash
git add .
git commit -a -m 'saving my slides'
git push
```
6. Wait a few minutes, then open your website, and add the file name of your slides to the URL. For example, if your jupyter file name is final_project.ipynb, and your website is:

http://qmisr.github.io/testing/

Then, to view your slides go to:

http://qmisr.github.io/testing/final_project.html

### Tip:
Once you successfully publish your slides, add a link to them from README.md to make it easier to reach the slides.
